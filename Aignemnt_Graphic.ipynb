```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from Similarity_Measures import Time_Series_Similarity, DTW, TDTW,LDTW,DDDTW,Eucledian,WDTW
from aeon.datasets import load_classification
```

```python
UCR_DATA, y = load_classification("CBF")
a=0
b=1
x = UCR_DATA[a][0]
x_show = UCR_DATA[a][0]-3
y = UCR_DATA[b][0]
display(len(x),len(y))
plt.figure(figsize=(15, 6))
plt.plot(x, color='blue')
plt.plot(x_show, color='green')
plt.plot(y, color='red')
```

```python
Result1=DTW(x, y)
# Result1=TDTW(x, y)
# Result1=LDTW(x, y, int(1.3*max(len(x),len(y))))
# Result1=WDTW(x, y,0.08)
# Result1=DDDTW(x, y,0)
# Result1=Euclidean(x, y)

# Result2=DTW(x, y)
Result2=TDTW(x, y)
# Result2=LDTW(x, y, int(1.3*max(len(x),len(y))))
# Result2=WDTW(x, y,0.08)
# Result2=DDDTW(x, y,0)
# Result2=Euclidean(x, y)
```

```python
plt.figure(figsize=(18, 8),facecolor='white')
plt.subplot(2, 1, 1)
plt.plot(x_show, color='red')
plt.plot(y, color='blue')
allignment_path = Result1.get_path()
# Draw lines connecting the aligned points
for (i, j) in allignment_path:
    plt.plot([i, j], [x_show[i], y[j]], 'r--', linewidth=0.5,color='Green')
print(len(allignment_path))
# Add labels and legend
plt.title(f"{Result1.get_name()} alignment, length: {Result1.len()}",fontsize=32)
plt.gca().set_xticks([])
plt.gca().set_yticks([])


plt.subplot(2, 1, 2)
plt.plot(x_show, color='red')
plt.plot(y, color='blue')
allignment_path = Result2.get_path()
# Draw lines connecting the aligned points
for (i, j) in allignment_path:
    plt.plot([i, j], [x_show[i], y[j]], 'r--', linewidth=0.5,color='Green')
print(len(allignment_path))
# Add labels and legend
plt.title(f"{Result2.get_name()} alignment, length: {Result2.len()}",fontsize=32)
plt.gca().set_xticks([])
plt.gca().set_yticks([])
```
